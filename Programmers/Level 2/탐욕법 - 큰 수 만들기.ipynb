      <h6 class="guide-section-title">문제 설명</h6>
      <div class="markdown solarized-dark"><p>어떤 숫자에서 k개의 수를 제거했을 때 얻을 수 있는 가장 큰 숫자를 구하려 합니다.</p>

<p>예를 들어, 숫자 1924에서 수 두 개를 제거하면 [19, 12, 14, 92, 94, 24] 를 만들 수 있습니다. 이 중 가장 큰 숫자는 94 입니다.</p>

<p>문자열 형식으로 숫자 number와 제거할 수의 개수 k가 solution 함수의 매개변수로 주어집니다. number에서 k 개의 수를 제거했을 때 만들 수 있는 수 중 가장 큰 숫자를 문자열 형태로 return 하도록 solution 함수를 완성하세요.</p>

<h5>제한 조건</h5>

<ul>
<li>number는 1자리 이상, 1,000,000자리 이하인 숫자입니다.</li>
<li>k는 1 이상 <code>number의 자릿수</code> 미만인 자연수입니다.</li>
</ul>

```python
def solution_1(number, k):
    import itertools
    return max([''.join(i) for i in set(itertools.combinations(number, len(number) - k))])
```

```python
def solution_2(number, k):
    res = ''
    res_len = len(number) - k
    while k:
        f = number[:k + 1]
        f_max = max(f)
        f_idx = f.index(f_max)
        res += f_max
        k -= f_idx
        number = number[f_idx + 1:]
        if k == 0:
            res += number
        if len(res) == res_len:
            break
    return res
```

```python
def solution_3(number, k):
    res = ''
    for i, j in enumerate(number):
        while len(res) and k and res[-1] < j:
            res = res[:-1]
            k -= 1
        if k == 0:
            res += number[i:]
            break
        res += j
    else:
        res = res[:-k]
    return res
```

```python
def solution_4(number, k):
    res = []
    for i, j in enumerate(number):
        while len(res) and k and res[-1] < j:
            res.pop()
            k -= 1
        if k == 0:
            res += number[i:]
            break
        res.append(j)
    else:
        res = res[:-k]
    return ''.join(res)
```

```python
def make_test(length=100):
    import random
    test_case = {}
    while len(test_case) < length:
        number = str(random.randrange(1, pow(10, 100)))
        k = random.randrange(1, len(number))
        if not len(test_case):
            k = len(number) - 5
        ans = solution_4(number, k)
        test_case[number, k] = ans
    return test_case
test_case = make_test()
for i, j in test_case.keys():
    print("{}\t{}".format(i, j))
```

```python
%timeit -n 1 -r 1 [solution_1(i, j) for i, j in list(test_case.keys())[:1]]
%timeit -n 100 -r 10 [solution_2(i, j) for i, j in list(test_case.keys())]
%timeit -n 100 -r 10 [solution_3(i, j) for i, j in list(test_case.keys())]
%timeit -n 100 -r 10 [solution_4(i, j) for i, j in list(test_case.keys())]
```

```python
%timeit -n 2 -r 2 solution_2("1450897413"*10000,50000)
%timeit -n 5 -r 2 solution_3("1450897413"*100000,500000)
%timeit -n 5 -r 2 solution_4("1450897413"*100000,500000)
```

solution_2 는 자리수가 작으면 빠르지만 자리수가 커질수록  
solution_3, solution_4가 빠르다.  
자리수가 커지면 커질수록 solution_4가 제일 좋은 성능을 보여준다.
